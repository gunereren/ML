## K-MEANS Model ve GÃ¶rselleÅŸtirme

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
kmeans
```

```python
k_fit = kmeans.fit(df.drop(columns=["species"]))
```

```python
k_fit.n_clusters
```

```python
kumeler = k_fit.labels_
```

```python
plt.scatter(x = df["petal_width"], y = df["sepal_width"], c= kumeler, s = 25, cmap="viridis")

plt.figure(figsize=(6, 6))
sns.scatterplot(x= "petal_width", y="sepal_width", hue="species", data=df);
```
